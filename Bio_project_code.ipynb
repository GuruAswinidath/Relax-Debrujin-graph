```python
#Debrujin graph

import networkx as nx
import matplotlib.pyplot as plt

graph = nx.Graph()
kmerk=[]              
suffix=[]
prefix=[]

def Construct(seq,k):
    for i in range(0,len(seq)-k+1):
        kmer = seq[i:i+k]
        kmerk.append(kmer)
    return kmerk

seq=input("Enter the dna sequencing")
k=int(input("Enter the k value"))
Construct(seq,k)
 
for i in kmerk:
    su=i[-(k-1):]
    pre=i[:(k-1)]
    suffix.append(su)
    prefix.append(pre)
    graph.add_edge(su,pre)

plt.figure(figsize=(20,20))

nx.draw_networkx(graph,node_size=1000,width=4)
"ACTTACGTACTTG,4"
"AAABBBA,3"
"CGATTCTAAG,3"
```

```python
#relax de bur3jin graph
import networkx as nx
import matplotlib.pyplot as plt

graph = nx.Graph()              
suffix=[]
prefix=[]
kmer1=[]

d={}

def count_kmers(seq, k):
    # Start with an empty dictionary
    counts = {}
    num_kmers = len(seq) - k + 1
    for i in range(num_kmers):
        kmer = seq[i:i+k]
        if kmer not in counts:
            counts[kmer] = 0
        counts[kmer] += 1
    return counts

def Construct(seq1,k):
   
    for i in range(0,len(seq1)-k+1):
        kmer = seq1[i:i+k]
        kmer1.append(kmer)

    for i in kmer1:
        su=i[-(k-1):]
        pre=i[:(k-1)]
        suffix.append(su)
        prefix.append(pre)
        d[pre]=su
        graph.add_edge(su,pre)
    return graph
    
seq1=input("Enter the dna 1st sequencing")
seq2=input("Enter the 2nd dna sequencing")
k=int(input("Enter the k value"))
Construct(seq1,k)
Construct(seq2,k)
    
print("The list of the kmers:",kmer1,"\n")
print("The lenght of the kmer list is:",len(kmer1),"\n")
print("The suffix list:",suffix,"\n")
print("The lenght of the suffix list is:",len(suffix),"\n")
print("The prefix list is:",prefix,"\n")
print("The length of the prefix is:",len(prefix),"\n")
print("The dictionary to construct graph is:",d,"\n")
print("The length of the dictinary",len(d),"\n")
print("The count of the kmers",count_kmers(seq1,k),count_kmers(seq2,k))
for i in d:
    print(d,)
plt.figure(figsize=(20,20))
nx.draw_networkx(graph,node_size=1000,width=4)

"CGATTCTAAGT,CGATTGTAAGT,3"
"ACTTACGTACTTG,ACTTACTTACT,4"
```
